### Dealing with text files

```python
# create a text file first and store it as "myfile.txt".
# opening the text file
myfile = open("myfile.txt")
```

```python
print(myfile)
```

```python
myfile = open("myfile.txt")
# to read the contents of the file
print(myfile.read())
```

```python
print(myfile.read())
```

```python
myfile.seek(0)
print(myfile.read())
```

```python
'''It is important to close the file once you are done so that other 
applications can use the file. To do so, call the close() method. '''
myfile.close()
```

```python
'''Instead of reading all the contents of the file at once, we can 
also read the file contents line by line using readlines() method.'''

myfile = open("myfile.txt")
print(myfile.readlines())
```

```python
myfile = open("myfile.txt")
for lines in myfile:
    print(lines.split()[0])
```

```python
myfile = open("myfile.txt", 'w+')
print(myfile.read())
```

```python
myfile.write("The file has been rewritten")
myfile.seek(0)
print(myfile.read())
```

```python
myfile = open("myfile.txt", 'a+')
myfile.write("\nThis is a new line")
myfile.seek(0)
print(myfile.read())
```

Using the with keyword, as shown below, we don't need to explicitly close the file. Rather, the below script opens the file, reads its contents, and then closes it automatically.

```python
with open("myfile.txt") as myfile:
    print(myfile.read())
```

# Dealing with text data
Here we see sentence tokenization and word tokenization using NLTK library.
Make sure that NLTK library is installed.

Sentence tokenization (also called sentence segmentation) is the problem 
of dividing a string of written language into its component sentences.

```python
import nltk
nltk.download('punkt') # punkt- tokenizer that divides a text into a list of sentences 
```

```python
text = "Welcome to MLFA class. Now we are about to learn sentence and word tokenizations. "
sentences = nltk.sent_tokenize(text)
sentences
```

```python
for sentence in sentences:
    print(sentence)
    print()
```

Word tokenization (also called word segmentation) is the problem of dividing a 
string of written language into its component words.
Let us use the sentences from the previous output and see how we can apply word tokenization on them. 
We can use the nltk.word_tokenize() function

```python
for sentence in sentences:
    words = nltk.word_tokenize(sentence)
    print(words)
    print()
```
